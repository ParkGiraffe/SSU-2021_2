### range 함수
  range(시작 , 끝 , 증감) 끝은 포함되지 않는다.  
    
  ex) 1~20까지 표시 >> range(1 , 21 , 1)  
  ex) 1,3,5,7,9 표시 >> range(1, 10 , 2)  
    
  range(1,10) = range(시작, 끝) ## 증감값을 안쓰면 디폴트 1로 인식  
  range(10) = range(끝) ## 증감값을 안쓰면 디폴트 1로 인식 , 시작값을 안쓰면 디포트 0  
  따로 자료형을 갖고 있지 않아서 list에 따로 담아줘야한다.  
  



 - 인덱스싱 a[1]
 - 슬라이싱 a[0:7]
 - 데이터 추가
   1) append : 끝에 추가  
   2) insert : 자리 지정 추가 listName.insert(id, item) | a.(1, 100)  
   3) extend : 확장(이어준다) 
  a = [1, 2, 3]  
  b = [4, 5, 6]  
  a.extend(b) # 1, 2, 3, 4, 5, 6  
  b는 그대로 [4, 5, 6]  

 - 데이터 삭제
   1) remove(item) ex) a.remove(111)  지정한 위치 값과 같은 값을 검색 후 '처음 값'을 삭제, 처음부터 순차 검색 후 만나면 삭제 및 함수 종료. 
   2) pop(id) ex) a.pop(0) 





### numpy
 - 왜 쓰나?
   1) array (수열)기능 - 행렬기능으로 n차 배열의 연산이 가능하다. (ex: 행렬 곱)  
    
  [[1, 2, 3, 4],
  [5, 6, 7, 8]]  
  
  list는 단차원 행렬으로 vector 형식,  
  다차원 행렬인 ndarray는 matrix라고 말함.  
  
   2) 실수 값 지원  
   arange표시 = 실수 연산 지원  
   list()는 실수 연산을 지원하지 않아서 numpy를 사용해야 한다.  
     
    a = list(range(0,4,0.5)) typeError 발생 - list는 float 지원 x -> numpy가 지원한다.  
    numpy - 수열기능 지원, 실수연산 지원.  
    a = np.arange(0, 4, 0.5)  
   
   
   3) 수학함수지원  
    sinAcitvatedArray = np.sin(ndArrayName) 사인값 계산.  



- np.arange(시작,종료,증가)
 - np.linspace(시작,종료,구분) # 등분



### 배열(ndArray)
1)장점
- 구현이 쉬움
- 인덱스가 있어 조회에서 빠른 성능을 보여줌 
- 연속된 메모리 공간에 할당되어 순차 접근에도 빠름
- 참조를 위한 추가 메모리 할당이 필요 없음

2)단점
- 삽입/삭제 시 뒤에 요소들을 이동해야 해서 비효율적
- 선언 시 지정한 크기 변경 불가
- 메모리 재사용 불가 (초기 사이즈만큼 할당이 되고, 데이터가 없더라도 메모리를 차지하고 있음)
 

### 리스트
1)장점
- 삽입/삭제 시 전후 노드의 참조 관계만 수정하면 되기 때문에 효율적
- 크기가 고정적이지 않음
- 메모리 재사용 가능 (삭제 시 해당 노드의 참조가 사라져 GC에 의해 가용 메모리로 전환)

2)단점
- 구현이 상대적으로 복잡
- 검색이 비효율적 (처음부터 순차적으로 검색해야 함)
- 참조를 위한 메모리가 필요  
  
결론
- 저장할 데이터의 개수가 정해져 있고 / 삽입, 삭제 작업이 적고 / 특정 위치의 데이터를 조회하는 작업이 많다면 배열
- 저장할 데이터의 개수가 미정이고 / 삽입, 삭제 작업이 많고 / 특정 위치 데이터를 조회하는 경우가 별로 없다면 리스트

```python
import matplotlib.pyplot as plt
import numpy as np
year = [1970, 1980, 1990, 2000, 2010] #연도
birth = [1006645, 862835, 649738, 634501, 470171] #출생아수

# 숫자를 문자로 바꾸는 방법2.
pos = np.arange(len(year)) # 하나의 변수 공간에 수열로 만들어서 적용시키면 됨.

plt.figure(figsize=(10, 7)) # 차트 종류를 정하는 plot() 위에 써야 한다. -> 차트 사이즈를 정한 후에 plot을 그린다고 보면 된다.

# plt.plot(year, birth, color = 'r', marker = 'o', linestyle = ':')
plt.bar(pos, birth)
plt.xticks(pos, year) # x의 눈금을 지정해줌. pos가 적힌 자리(0, 1, 2, 3, 4)를 year로 변경시킴 -> 갯수가 다르면 오류를 냄.
plt.xlabel('year')
plt.ylabel('birth')
plt.ylim(300000, 1200000)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 숫자를 문자로 바꾸는 방법3. 모두 스트링으로
year = ['1970', '1980', '1990', '2000', '2010'] #연도
birth = [1006645, 862835, 649738, 634501, 470171] #출생아수

pos = np.arange(len(year)) # 하나의 변수 공간에 수열로 만들어서 적용시키면 됨.

plt.figure(figsize=(10, 7)) # 차트 종류를 정하는 plot() 위에 써야 한다. -> 차트 사이즈를 정한 후에 plot을 그린다고 보면 된다.

# plt.plot(year, birth, color = 'r', marker = 'o', linestyle = ':')
plt.bar(pos, birth, alpha=0.5, label = 'bar') # alpha: 투명도
plt.plot(pos, birth, color = 'r', label = 'plot') # alpha: 투명도
plt.xticks(pos, year) # x의 눈금을 지정해줌. pos가 적힌 자리(0, 1, 2, 3, 4)를 year로 변경시킴
plt.xlabel('year')
plt.ylabel('birth')
plt.ylim(300000, 1200000)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

# pandas
# f = read_csv()

f = open('/Users/yosep/OneDrive - 숭실대학교 - Soongsil University/과제/SSU2021-2_git/AI와데이터사회/6주차/a.csv', encoding='CP949') # 저장 만 함. 따로 읽는 과정이 필요함.
d = csv.reader(f) # 저장한 csv를 읽는 과정.

next(d) # 첫째 줄을 버림 == 열 제목 제거

address1 = [] # 월계3동의 0세부터 n세 까지의 인구를 담아놓는 list
address2 = []

add1 = input('첫 번째 지역 입력 : ')
add2 = input('두 번째 지역 입력 : ')

for k in d: # csv 파일을 레코드 단위로 가져옴
    if add1 in k[0] : # 줄에서 0번 째 칸의 요소를 검사함.
        for i in range(3, 104, 1) : # 0번은 지역 이름, 1번과 2번은 합계가 담겨 있음. col은 총 104개가 존재.
            address1.append(int(k[i].replace(',' , ''))) # csv 안의 요소는 string으로 가져오기 때문에, int 형으로 변환이 필요.
                                                           # replace로 쉼표 제거.
    if add2 in k[0] :
        for i in range(3, 104, 1) :
            address2.append(int(k[i].replace(',' , '')))


plt.plot(address1, label = add1)
plt.plot(address2, label = add2)

plt.rc('font', family='AppleGothic') # window에서 실행하기
plt.legend()
plt.show()

# 역삼1동에서 에러가 남.
# invalid literal for int() with base 10: '1,144'
# --> 인구수가 천 단위가 넘어가서 쉼표(',')가 들어가기 때문이다.
# 쉼표가 있으면 숫자로 인식 불가. int()로도 불가능.
# replace를 이용해서 쉼표를 없애야 한다. replace(old_str, new str)
```
