```
from sklearn import tree
```

```
x = [[0,0], [1,1], [1,0]]
y = [0, 1, 0] 
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)
```

```
clf.predict([[1.,0.,]])    # 새로운 데이터에 대한 예측 
```

```
clf.predict([[0.7,0.6,]])   # 새로운 데이터에 대한 예측
```

```
from sklearn.tree import export_graphviz
export_graphviz(clf, 
                out_file='tree.dot', 
                class_names=['음성', '양성'], 
                impurity=False, 
                filled=True)
```

```
import graphviz
with open('tree.dot') as f : 
  dot_graph = f.read()
display(graphviz.Source(dot_graph))
```
