```python
f = sns.load_dataset('tips')
f # 개발자가 쉽게 약식으로 확인하는 용도
# print(f) - 개발자가 정식으로 프로그램 사용자에게 해당 변수를 보여주기 위해 사용한다.
```

```python
f.info()  #실제 데이터는 안 보여주고 칼럼만 보여줌.
#index는 col 취급 안 함.
```

```python
f.describe()
#분석이 가능한 데이터의 통계를 보여준다.
#분석이 가능하다 => 수치형 데이터.
# 갯수 count
# 평균 mean
# 표준편차 std
# 최소값 min
# 분위 %
# 최대값 max
```

```python
#슬라이스
f[50:70]
# ':' - a에서부터 b이전까지 표시 => b에서 실행 종료.
```

```python
# 시각화 기능
# rel : 점
# plot : 차트
sns.relplot(data=f, x='total_bill', y="tip")
sns.relplot(data=f, x='total_bill', y="tip", col="time", hue="smoker", style="smoker", size="size")

# col : 'time'을 기준으로 차트를 만들겠다.
# size : 몇 명(size)이 같이 왔냐에 따라 점의 크기를 다르게 한다.
# hue : 담배 피는 사람이면 주황색
```
